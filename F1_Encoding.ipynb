```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/WithConfidence.csv'
df = pd.read_csv(file_path)

pd.set_option('display.max_rows', None)
vc = df['Driver'].value_counts(20)
print(vc)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Weather'] = le.fit_transform(df['Weather'])
df.head(40)
```

```python
df['Winner']=df['Position'].apply(lambda x: 1 if x==1 else 0)
df.head(20)
```

```python
df['Track'].value_counts()
df['Driver'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Track_encoded'] = le.fit_transform(df['Track'])
df['Country_encoded'] = le.fit_transform(df['Country'])
df['Driver_encoded'] = le.fit_transform(df['Driver'])
df['Team_encoded'] = le.fit_transform(df['Team'])
df = df[['Year', 'Track', 'Track_encoded', 'Country', 'Country_encoded', 'Position', 'Driver', 'Driver_encoded',
         'Team', 'Team_encoded', 'Starting Grid', 'Points', 'Pitstop Time', 'Weather', 'avg_confidence_score']]
df.head(20)
```

```python
from google.colab import files

df.to_csv('Label_Encoded.csv', index=False)

files.download('Label_Encoded.csv')
```
