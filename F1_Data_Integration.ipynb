```python
import pandas as pd
import numpy as np
```

```python
df_constructor = pd.read_csv('/content/df_constructor.csv')
df_pitstop = pd.read_csv('/content/df_pitstop.csv')
df_race_result = pd.read_csv('/content/df_race_result.csv')
df_weather = pd.read_csv('/content/df_weather.csv')
```

```python
df_race_result.head()
```

```python
df_pitstop.rename(columns={'Circuit': 'Track'}, inplace=True)
df_pitstop.rename(columns={'Driver Name': 'Driver'}, inplace=True)
df_weather.rename(columns={'Circuit': 'Track'}, inplace=True)
```

```python
df_pitstop.head()
```

```python
df_weather.head()
```

```python
df = df_race_result.merge(df_pitstop, on=['Year', 'Track','Country','Driver'], how='left')
```

```python
df = df.merge(df_weather, on=['Year', 'Track'], how='left')
```

```python
from google.colab import files
df.to_csv('Integrated_df.csv', index=False)
files.download('Integrated_df.csv')
```
