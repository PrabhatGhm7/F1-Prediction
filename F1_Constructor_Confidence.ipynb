```python
import pandas as pd
```

```python
df = pd.read_csv('/content/df_constructor.csv')
```

```python

# 1. Compute the max position and max points per year
df['max_position_year'] = df.groupby('Year')['Position'].transform('max')
df['max_points_year'] = df.groupby('Year')['Points'].transform('max')

# 2. Define weights (alpha and beta)
alpha = 0.5
beta = 0.5

# 3. Calculate confidence for each row
df['confidence_score'] = (
    alpha * (1 - (df['Position'] - 1) / (df['max_position_year'] - 1)) +
    beta * (df['Points'] / df['max_points_year'])
)

# 4. Optional: round the confidence score
df['confidence_score'] = df['confidence_score'].round(3)

# 5. Drop the helper columns if you like
df.drop(['max_position_year', 'max_points_year'], axis=1, inplace=True)

```

```python
# View the updated DataFrame
print(df.head(20))
```

```python
# Group by Constructor and compute the average confidence score
df_avg_confidence = df.groupby('Constructor', as_index=False)['confidence_score'].mean()
df_avg_confidence.rename(columns={'confidence_score': 'avg_confidence_score'}, inplace=True)
print(df_avg_confidence)
```

```python
from google.colab import files
df_avg_confidence.to_csv('Constructor_Confidence.csv', index=False)
files.download('Constructor_Confidence.csv')
df_confidence = pd.read_csv('/content/df_constructor_confidence.csv')
df_no_duplicate = pd.read_csv('/content/NoDuplicate.csv')
```
