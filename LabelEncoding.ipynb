```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = '/content/drive/MyDrive/WithConfidence.csv'
df = pd.read_csv(file_path)

pd.set_option('display.max_rows', None)
vc = df['Driver'].value_counts(20)
print(vc)
```

```python
df['Winner']=df['Position'].apply(lambda x: 1 if x==1 else 0)
df
```

```python
df['Track'].value_counts()
df['Driver'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Track'] = le.fit_transform(df['Track'])
df['Country'] = le.fit_transform(df['Country'])
df['Driver'] = le.fit_transform(df['Driver'])
df['Team'] = le.fit_transform(df['Team'])
df
```
